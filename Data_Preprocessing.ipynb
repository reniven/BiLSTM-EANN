```python
df = pd.read_csv("set_images.txt",  sep="\t", )
```

```python
m, n = df.shape
```

```python
df.iloc[0]
```

```python
# download and save images as .jpg
for i in range(m):
    file_name = df.iloc[i]['image_id']
    url = df.iloc[i]['image_url']
    try:
        res = requests.get(url, stream = True)
    except requests.ConnectionError as err:
        print(f'{file_name} Couldn\'t be retrieved')
        continue
#     print(file_name)
    if res.status_code != 200:
        with open('images/'+file_name+'.jpg','wb') as f:
            shutil.copyfileobj(res.raw, f)
    #     print('Image sucessfully Downloaded: ',file_name)
    else:
        print(f'{file_name} Couldn\'t be retrieved')
```

```python
# print the file_name with dead links
for i in range(m):
    file_name = df.iloc[i]['image_id']
    url = df.iloc[i]['image_url']
    try:
        res = requests.get(url, stream = True)
    except requests.ConnectionError as err:
        print(file_name)
        continue
    if res.status_code != 200:
        print(file_name)
```

```python
df_post_features = pd.read_csv("mediaeval2016/devset/dataset_features/post_features.txt")
df_user_features = pd.read_csv("mediaeval2016/devset/dataset_features/user_features.txt")
```

```python
# remove url links from posts
df_posts['post_text_remove_url'] = \
                df_posts['post_text'].str.replace(r'(?:(?:https?|ftp):\/\/)[\w/\-?=%.]+\.[\w/\-&?=%.]+', '')
```

```python
# merge(left join) df_posts, df_posts_features, df_user_features based on the "post_id" column
df = df_posts.merge(df_post_features,on='post_id',how='left').merge(df_user_features,on='post_id',how='left')
```

```python
df.columns=df.columns.str.strip()
```

```python
# drop unnecssary columns=['post_text', 'num_mentions', 'num_retweets'] 
df = df.drop(columns=['post_text', 'num_mentions', 'num_retweets'])
```
